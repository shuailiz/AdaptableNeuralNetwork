```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import os

from PIL import Image
```

```python
with open('../datasets/cifar-10-batches-py/data_batch_1', 'rb') as fo:
    dict = pickle.load(fo, encoding='bytes')
dict
```

```python
dict.keys()
```

```python
ind = 1981
print(dict[b'filenames'][ind])
r_channel = np.reshape(dict[b'data'][ind][:1024], (32, 32))
g_channel = np.reshape(dict[b'data'][ind][1024:2048], (32, 32))
b_channel = np.reshape(dict[b'data'][ind][2048:], (32, 32))

image = np.zeros((32, 32, 3))
image[:, :, 0] = r_channel / 255
image[:, :, 1] = g_channel / 255
image[:, :, 2] = b_channel / 255

plt.imshow(image)
```
